```python
import requests
import bs4
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)
```

```python
def read_data(url):
    """
    read in the data from the url.
    ---------------
    Parameters:
    ---------------
    
    Inputs: 
        url (string): the url of interest 

    Outputs: 
        resulted (HTML): the HTML code to search through
    
    """
    r = requests.get(url)
    print(r)
    resulted = bs4.BeautifulSoup(r.text, 'html')
    return(resulted)
```

```python
url = "https://collegefootballnetwork.com/highest-paid-college-football-coaches/"
webpage_html = read_data(url)
```

```python
raw_data = webpage_html.find_all('p')[5:-1]

refined_data = []
raw_data_split = [str(i).split('<br/>') for i in raw_data]

for lst in raw_data_split:
    if 'Listed below are the top-paid college' not in lst[0]:
        refined_data += lst
```

```python
salaries = []
for salary in [i.split(')')[-1].split('$')[-1].replace('</p>', '') for i in refined_data]:
    try:
        salaries.append(float(salary.replace(',','')))
    except:
        salaries.append(np.nan)



schools = []
for school in [i.split(':')[0].split(',')[-1][1:] for i in refined_data]:
    schools.append(school)


coaches_salaries = pd.DataFrame({'school':schools, 'salary':salaries}).dropna().sort_values(['school', 'salary']).drop_duplicates('school', keep='last').reset_index(drop=1)

# replace Texas A&amp;M  with Texas A&M
coaches_salaries.loc[coaches_salaries['school'] == 'Texas A&amp;M', 'school'] = 'Texas A&M'
```
